```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Customer Churn.csv')
df.head()
```

```python
df.info()
```

```python
df["TotalCharges"]=df["TotalCharges"].replace(" ",0)
df["TotalCharges"]=df["TotalCharges"].astype("float")
```

```python
df.info()
```

```python
df.isnull().sum().sum()
```

```python
df["customerID"].duplicated().sum()
```

```python
def cons(value):
    if value==1:
        return "yes"
    else:
        return "no"
df["SeniorCitizen"]=df["SeniorCitizen"].apply(cons)
```

```python

sns.countplot(x='Churn',data=df)
plt.show()
```

```python
gb =df.groupby('Churn').agg({"Churn":'count'})
plt.pie(gb['Churn'],labels =gb.index,autopct="%1.2f%%")
plt.show()
```

```python
plt.figure(figsize=(3,3))
sns.countplot(x='gender',data=df,hue="Churn",)
plt.title("Churn By Gender")
plt.show()
```

```python
plt.figure(figsize=(3,3))
sns.countplot(x='SeniorCitizen',data=df,hue="Churn")
plt.title("Churn By Gender")
plt.show()
```

```python
plt.figure(figsize=(4,4))
ax= sns.countplot(x='SeniorCitizen',data=df)
ax.bar_label(ax.containers[0])
plt.title("Churn By Gender")
plt.show()
```

```python
# Create a crosstab for stacked bar plot and normalize it to get percentages
ct = pd.crosstab(df['SeniorCitizen'], df['Churn'], normalize='index') * 100

# Plot the stacked bar plot
ax = ct.plot(kind='bar', stacked=True, figsize=(5,5))

# Add percentages inside the bars
for p in ax.patches:
    width = p.get_width()    # Get bar width
    height = p.get_height()  # Get bar height
    x = p.get_x() + width / 2
    y = p.get_y() + height / 2
    ax.annotate(f'{height:.1f}%', (x, y), ha='center', va='center', color='white', fontsize=10)

plt.title("Churn By Senior Citizen (Percentage)")
plt.xlabel("Senior Citizen")
plt.ylabel("Percentage (%)")
plt.legend(title="Churn", loc='upper right')
plt.show()
```

```python
features = ['PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 
            'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies']

# Number of subplots
n_features = len(features)

# Set up the matplotlib figure and axes
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(20, 22))

# Flatten the axes array for easier iteration
axes = axes.flatten()

# Create a countplot for each feature
for i, feature in enumerate(features):
    sns.countplot(data=df, x=feature, ax=axes[i],hue=df["Churn"])
    axes[i].set_title(f'Count of {feature}')
    axes[i].set_xlabel(feature)
    axes[i].set_ylabel('Count')

# Adjust the layout to avoid overlapping
plt.tight_layout()

# Show the plots
plt.show()
```

#Customers without value-added services like OnlineSecurity, Backup, and TechSupport are more likely to churn. Fiber optic internet users show higher churn rates, while customers without internet service churn less. PhoneService and MultipleLines don't significantly affect churn. Streaming services have a balanced churn impact. Generally, customers lacking optional services tend to churn more
